```python
birds_ebird = set(df_ebird['Avibase.ID2'].unique())
birds_blife = set(df_blife['Avibase.ID1'].unique())
common_birds = birds_ebird.intersection(birds_blife)
```

```python
len(common_birds)
```

```python
df_blife_common = df_blife[df_blife['Avibase.ID1'].isin(common_birds)].copy()
df_ebird_common = df_ebird[df_ebird['Avibase.ID2'].isin(common_birds)].copy()
```

```python
df_blife_common['Avibase.ID'] = df_blife_common['Avibase.ID1'].copy()
df_ebird_common['Avibase.ID'] = df_ebird_common['Avibase.ID2'].copy()
```

```python
common_columns = list(set(df_blife_common.columns).intersection(set(df_ebird_common.columns)))
```

```python
len(common_columns)
```

```python
df_blife_common = df_blife_common[common_columns].copy()
df_ebird_common = df_ebird_common[common_columns].copy()
```

```python
df_blife_common.shape
```

```python
df_ebird_common.shape
```

```python
df_blife_common['Avibase.ID'].nunique(), df_ebird_common['Avibase.ID'].nunique()
```

```python
df_blife_common.sort_values(by=['Avibase.ID'], inplace=True)
df_ebird_common.sort_values(by=['Avibase.ID'], inplace=True)
```

```python
df_ebird_common.reset_index(drop=True, inplace=True)
df_blife_common.reset_index(drop=True, inplace=True)
```

```python
df_common_equality = df_blife_common == df_ebird_common
```

```python
df_common_equality.sum().sort_values(ascending=False)
```

```python
df_common_merge = df_blife_common.merge(df_ebird_common, how='inner', on='Avibase.ID', suffixes=['_blife', '_ebird' ])
```

```python
for col in df_blife_common.columns:
    if col == 'Avibase.ID':
        continue
    df_common_merge[col+'_equal'] = df_common_merge[col+'_blife'] == df_common_merge[col+'_ebird']
```

```python
column_to_investigate = ['Reference.species']
```

```python

def generate_columns_select(column_to_investigate):
    column_to_select = []
    for col in column_to_investigate:
        col1 = col + '_blife'
        col2 = col + '_ebird'
        col3 = col + '_equal'
        column_to_select.append(col1)
        column_to_select.append(col2)
        column_to_select.append(col3)
    column_to_select.append('Avibase.ID')
    return column_to_select
```

```python
column_to_select = generate_columns_select(['Reference.species'])
df_common_merge[df_common_merge['Reference.species_equal'] == False].sort_values(by='Reference.species_ebird')[column_to_select]
```

```python
cols = ['Inference'           
,'Male'                
,'Trophic.Level'       
,'Total.individuals'   
,'Trophic.Niche'       
,'Mass.Source'         
,'Kipps.Distance']
```
