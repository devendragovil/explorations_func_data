```python
df_common_merge = df_blife_common.merge(df_ebird_common, how='inner', on='Avibase.ID', suffixes=['_blife', '_ebird' ])
```

```python
for col in df_blife_common.columns:
    if col == 'Avibase.ID':
        continue
    df_common_merge[col+'_equal'] = df_common_merge[col+'_blife'] == df_common_merge[col+'_ebird']
```

```python
column_to_investigate = ['Reference.species']
```

```python

def generate_columns_select(column_to_investigate):
    column_to_select = []
    for col in column_to_investigate:
        col1 = col + '_blife'
        col2 = col + '_ebird'
        col3 = col + '_equal'
        column_to_select.append(col1)
        column_to_select.append(col2)
        column_to_select.append(col3)
    column_to_select.append('Avibase.ID')
    return column_to_select
```

```python
column_to_select = generate_columns_select(['Reference.species'])
df_common_merge[df_common_merge['Reference.species_equal'] == False].sort_values(by='Reference.species_ebird')[column_to_select]
```

```python
numerical_columns = [
    'Female_ebird', 'Total.individuals_ebird', 'Wing.Length_ebird',
       'Tarsus.Length_ebird', 'Beak.Depth_ebird', 'Complete.measures_ebird',
       'Beak.Length_Culmen_ebird', 'Habitat.Density_ebird', 'Migration_ebird',
       'Beak.Length_Nares_ebird', 'Hand-Wing.Index_ebird', 'Male_ebird',
       'Tail.Length_ebird',
       'Kipps.Distance_ebird', 'Beak.Width_ebird', 'Secondary1_ebird', 'Mass_ebird'
]
```

```python
X = df_common_merge_shortlisted_ebird[numerical_columns]
```

```python
from sklearn.decomposition import PCA
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std = StandardScaler()
pca = PCA()
```

```python
X_std = std.fit_transform(X)
```

```python
pca.fit(X_std)
```

```python
pca.explained_variance_ratio_
```

```python
scree = np.cumsum(pca.explained_variance_ratio_)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(scree)
plt.title("Scree Plot")
plt.ylabel("Explained Variance Ratio")
plt.xlabel("Number of Eigenvectors")
```
