```python
import geopandas, shapely, fiona
```

```python
state_shapefile = geopandas.read_file('../data/States_shapefile/States_shapefile.shp')
```

```python
nevada_polygon = state_shapefile.loc[28, 'geometry']
```

```python
pt = shapely.geometry.Point(-117.0026164111954, 39.459391567353045)
```

```python
nevada_polygon.contains(pt)
```

```python
def reverse_geocode_isNV(long, lat):
    pt = shapely.geometry.Point(long, lat)
    return nevada_polygon.contains(pt)
```

```python
df_iNat_filtered['state_is_NV'] = df_iNat_filtered[['Longitude', 'Latitude']].apply(
    lambda x: reverse_geocode_isNV(x['Longitude'], x['Latitude']),
    axis = 1
)
```

```python
df_iNat_filtered
```

```python
df_iNat_filtered['state_is_NV'].value_counts()
```

```python
df_iNat_filtered2 = df_iNat_filtered[df_iNat_filtered['state_is_NV'] == False]
```

```python
df_iNat_filtered2
```

```python
df_iNat_filtered2['state_is_NV'].value_counts()
```

```python
map_1 = KeplerGl()
map_1.add_data(data=df_iNat_filtered2[['Latitude', 'Longitude']], name='iNat')
```
