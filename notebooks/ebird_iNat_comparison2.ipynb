```python
def circular_smoothing(values, a=0.5, b=0.25):
    if a+b+b != 1:
        raise ValueError("a+b+b must be 1")
    to_return = []
    for i in range(len(values)):
        current_val = values[i]
        prev_val = values[i-1]
        if i == 11:
            next_val = values[0]
        else:
            next_val = values[i+1]
        val_to_return = (a*current_val) + (b*prev_val) + (b*next_val)
        to_return.append(val_to_return)
    
    return to_return
```

```python
df_iNat_birds_2_csmth = []
```

```python
for row in df_iNat_birds_2.values:
    row_ls = list(row)
    output = circular_smoothing(row_ls)
    df_iNat_birds_2_csmth.append(output)
    
```

```python
df_iNat_birds_2_csmth2= pd.DataFrame(np.array(df_iNat_birds_2_csmth))
```
